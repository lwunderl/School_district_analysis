```python
import pandas as pd
import os
```

```python
schools_csv = os.path.join("Resources", "schools_complete.csv")
students_csv = os.path.join("Resources", "students_complete.csv")
```

```python
schools = pd.read_csv(schools_csv)
students = pd.read_csv(students_csv)
```

```python
df_merged_schools_students = pd.merge(schools, students, how="left", on=["school_name", "school_name"])
```

```python
complete_student_count = len(df_merged_schools_students["student_name"])
```

```python
complete_school_count = len(df_merged_schools_students["school_name"].unique())
```

```python
complete_budget = df_merged_schools_students.groupby(["school_name"]).mean()["budget"].sum()
```

```python
complete_average_math = float(df_merged_schools_students["math_score"].mean())
```

```python
complete_average_reading = float(df_merged_schools_students["reading_score"].mean())
```

```python
complete_passing_math_count = df_merged_schools_students[(df_merged_schools_students["math_score"] >= 70)].count()["student_name"]
complete_passing_math_rate = complete_passing_math_count / float(complete_student_count) * 100
```

```python
complete_passing_reading_count = df_merged_schools_students[(df_merged_schools_students["reading_score"] >= 70)].count()["student_name"]
complete_passing_reading_rate = complete_passing_reading_count / float(complete_student_count) * 100
```

```python
complete_passing_math_reading_count = df_merged_schools_students[
    (df_merged_schools_students["math_score"] >= 70) & (df_merged_schools_students["reading_score"] >= 70)
].count()["student_name"]
complete_passing_math_reading_rate = complete_passing_math_reading_count /  float(complete_student_count) * 100
```

```python
df_complete_summary = pd.DataFrame({
    "Total Schools": [complete_school_count],
    "Total Students": [complete_student_count],
    "Total Budget": [complete_budget],
    "Average Math Score": [complete_average_math],
    "% Passing Math": [complete_passing_math_rate],
    "Average Reading Score": [complete_average_reading],
    "% Passing Reading": [complete_passing_reading_rate],
    "% Overall Passing": [complete_passing_math_reading_rate],
})

df_complete_summary["Total Students"] = df_complete_summary["Total Students"].map("{:,}".format)
df_complete_summary["Total Budget"] = df_complete_summary["Total Budget"].map("${:,.2f}".format)
df_complete_summary["% Passing Math"] = df_complete_summary["% Passing Math"].map("{:,.2f}%".format)
df_complete_summary["% Passing Reading"] = df_complete_summary["% Passing Reading"].map("{:,.2f}%".format)
df_complete_summary["% Overall Passing"] = df_complete_summary["% Overall Passing"].map("{:,.2f}%".format)
df_complete_summary["Average Math Score"] = df_complete_summary["Average Math Score"].map("{:,.2f}".format)
df_complete_summary["Average Reading Score"] = df_complete_summary["Average Reading Score"].map("{:,.2f}".format)

df_complete_summary
```

```python
school_type = schools.set_index(["school_name"])["type"]
```

```python
per_school_count = schools.set_index(["school_name"])["size"]
```

```python
per_school_budget = df_merged_schools_students.groupby(["school_name"]).mean()["budget"]
```

```python
per_school_capita = per_school_budget / per_school_count
```

```python
per_school_ave_math = df_merged_schools_students.groupby(["school_name"]).mean()["math_score"]
```

```python
per_school_ave_reading = df_merged_schools_students.groupby(["school_name"]).mean()["reading_score"]
```

```python
student_passing_math = (df_merged_schools_students[df_merged_schools_students["math_score"] >= 70])
per_school_passing_math = student_passing_math.groupby(["school_name"]).count()["student_name"]
```

```python
student_passing_reading = (df_merged_schools_students[df_merged_schools_students["reading_score"] >= 70])
per_school_passing_reading = student_passing_reading.groupby(["school_name"]).count()["student_name"]
```

```python
student_passing_math_reading = df_merged_schools_students[(df_merged_schools_students["math_score"] >= 70) & (df_merged_schools_students["reading_score"] >= 70)]
per_school_passing_math_reading = student_passing_math_reading.groupby(["school_name"]).count()["student_name"]
```

```python
per_school_passing_math_rate = per_school_passing_math / per_school_count * 100
per_school_passing_reading_rate = per_school_passing_reading / per_school_count * 100
per_school_passing_math_reading_rate = per_school_passing_math_reading / per_school_count * 100
```

```python
df_per_school_summary = pd.DataFrame({
    "School Type": school_type,
    "Total Students": per_school_count,
    "Total School Budget": per_school_budget,
    "Per Student Budget": per_school_capita,
    "Average Math Score": per_school_ave_math,
    "% Passing Math": per_school_passing_math_rate,
    "Average Reading Score": per_school_ave_reading,
    "% Passing Reading": per_school_passing_reading_rate,
    "% Overall Passing": per_school_passing_math_reading_rate
})

df_per_school_summary["Total Students"] = df_per_school_summary["Total Students"].map("{:,.0f}".format)
df_per_school_summary["Total School Budget"] = df_per_school_summary["Total School Budget"].map("${:,.2f}".format)
df_per_school_summary["Per Student Budget"] = df_per_school_summary["Per Student Budget"].map("${:,.2f}".format)
df_per_school_summary["Average Math Score"] = df_per_school_summary["Average Math Score"].map("{:.2f}".format)
df_per_school_summary["Average Reading Score"] = df_per_school_summary["Average Reading Score"].map("{:.2f}".format)
df_per_school_summary["% Passing Math"] = df_per_school_summary["% Passing Math"].map("{:.2f}%".format)
df_per_school_summary["% Passing Reading"] = df_per_school_summary["% Passing Reading"].map("{:.2f}%".format)
df_per_school_summary["% Overall Passing"] = df_per_school_summary["% Overall Passing"].map("{:.2f}%".format)

df_per_school_summary
```

```python
ninth_graders = df_merged_schools_students[(df_merged_schools_students["grade"] == "9th")]
tenth_graders = df_merged_schools_students[(df_merged_schools_students["grade"] == "10th")]
eleventh_graders = df_merged_schools_students[(df_merged_schools_students["grade"] == "11th")]
twelfth_graders = df_merged_schools_students[(df_merged_schools_students["grade"] == "12th")]
```

```python
ninth_graders_ave_math = ninth_graders.groupby(["school_name"]).mean()["math_score"]
tenth_graders_ave_math = tenth_graders.groupby(["school_name"]).mean()["math_score"]
eleventh_graders_ave_math = eleventh_graders.groupby(["school_name"]).mean()["math_score"]
twelfth_graders_ave_math = twelfth_graders.groupby(["school_name"]).mean()["math_score"]
```

```python
ninth_graders_ave_reading = ninth_graders.groupby(["school_name"]).mean()["reading_score"]
tenth_graders_ave_reading = tenth_graders.groupby(["school_name"]).mean()["reading_score"]
eleventh_graders_ave_reading = eleventh_graders.groupby(["school_name"]).mean()["reading_score"]
twelfth_graders_ave_reading = twelfth_graders.groupby(["school_name"]).mean()["reading_score"]
```

```python
math_score_ave_by_grade = pd.concat([
    ninth_graders_ave_math,
    tenth_graders_ave_math,
    eleventh_graders_ave_math,
    twelfth_graders_ave_math],
    axis=1)
math_score_ave_by_grade
```

```python
reading_score_ave_by_grade = pd.concat([
    ninth_graders_ave_reading,
    tenth_graders_ave_reading,
    eleventh_graders_ave_reading,
    twelfth_graders_ave_reading],
    axis=1)
reading_score_ave_by_grade
```
